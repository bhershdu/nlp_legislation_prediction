```python
import os
import pandas as pd
import numpy as np
from tsne_torch import TorchTSNE as TSNE
import fnmatch
import csv
```

```python
token_path = os.path.join(os.getcwd(),"..","data","tokenized")
```

```python
number_to_load = 1987
```

```python
data_file = "resample-data.txt"
label_file = "resample-label.txt"
```

```python
with open(data_file, 'w') as d, open(label_file, 'w') as l:
    data_writer = csv.writer(d, dialect='unix', delimiter=" ", quoting=csv.QUOTE_NONE)
    label_writer = csv.writer(l, dialect='unix', delimiter=" ", quoting=csv.QUOTE_NONE)
    file_count = 0
    for root, dirs, files in os.walk(token_path):
        for f in files:
            if fnmatch.fnmatch(f, "*shrunk*") and not fnmatch.fnmatch(f, "*512*"):
                file_count += 1
                if file_count <= number_to_load:
                    print(f'loading {f}')
                    df = pd.read_pickle(os.path.join(root, f), compression="gzip")
                    encoding = np.array(df["input_shrunk"]).tolist()
                    party_label = df["party"][0]
#                    l_array = np.array(df["party"])
#                    i_index = np.argmax(l_array, axis=0)
#                    i_index_str = str(i_index)
                    data_writer.writerow(encoding)
                    label_writer.writerow([party_label])
```
