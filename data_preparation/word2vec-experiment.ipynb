Want to see how word2vec would work on my data

```python
from gensim.models import Word2Vec
from gensim.test.utils import common_texts
import os
import json
```

```python
model = Word2Vec(sentences=common_texts, vector_size=100, window=5, min_count=1, workers=4)
```

```python
vector = model.wv['computer']
```

```python
vector
```

```python
sample_file = os.path.join(os.getcwd(),"..","data","extracted", "summary_bill_text_1729_1287633.json")
```

```python
with open(sample_file, 'r') as f:
    data_json = json.load(f)
```

```python
words = data_json["text"].split(" ")
```

```python
words
```

```python
model.train(words, total_examples=1, epochs=1 )
```

```python
model.wv[words[0]]
```

```python
import fnmatch
```

```python
word_set = None
for root, dirs, files in os.walk(os.path.join(os.getcwd(), "..", 'data', "extracted")):
    for f in files:
        if fnmatch.fnmatch(f, "summary_bill_text*.json"):
            with open(os.path.join(root, f)) as d_file:
                data = json.load(d_file)
                words = set(data["text"].split(" "))
                if word_set is None:
                    word_set = words
                else:
                    word_set = word_set.union(words)
```

```python
len(word_set)
```

```python
model.train(word_set, total_examples=len(word_set), epochs=20 )
```

```python
model.wv[word_set]
```
